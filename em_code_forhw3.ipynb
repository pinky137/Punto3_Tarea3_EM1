```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
L = 0.1           # Longitud de cada lado del cuadrado (metros)
N = 101           # Número de puntos en cada dirección
h = L / (N - 1)   # Espaciado de la cuadrícula (0.001 metros)
epsilon0 = 8.85e-12  # Permisividad del vacío (F/m)

# Inicializar matriz del potencial
V = np.zeros((N, N))

# Condiciones de contorno
V[:, 0] = 0        # Lado izquierdo (0 V)
V[:, -1] = 0       # Lado derecho (0 V)
V[0, :] = 0        # Base (0 V)
V[-1, :] = 5.0     # Tapa superior (5 V)

# Iteraciones para calcular el potencial usando el método de diferencias finitas
max_iter = 2000   # Número máximo de iteraciones
tolerance = 1e-6   # Tolerancia para la convergencia

for _ in range(max_iter):
    V_new = np.copy(V)
    for i in range(1, N - 1):
        for j in range(1, N - 1):
            V_new[i, j] = (1/4) * (V[i, j + 1] + V[i, j - 1] + V[i + 1, j] + V[i - 1, j])

    # Comprobar la convergencia
    if np.max(np.abs(V_new - V)) < tolerance:
        V = V_new
        break
    else:
        V = V_new

# Crear una gráfica de calor para visualizar el potencial
plt.imshow(V, cmap='viridis', extent=[0, L, 0, L], origin='lower', aspect='auto')
plt.colorbar(label='Potencial eléctrico (V)')
plt.title('Potencial Eléctrico en el Cuadrado')
plt.xlabel('Posición en x (m)')
plt.ylabel('Posición en y (m)')
plt.show()

```
